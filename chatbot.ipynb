```python
from google.colab import files
import pandas as pd
import json
import datetime
```

```python
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
print("Dataset loaded successfully!")
df.head()
```

```python
df.dropna(inplace=True)  # Remove empty rows

def clean_text(text):
    return text.strip() if isinstance(text, str) else text

df = df.applymap(clean_text)
print("Dataset cleaned!")
```

```python
responses = {}
for _, row in df.iterrows():
    category = row["Category"]
    query = row["Query"].lower()
    response = row["Response"]
    responses[query] = (category, response)
```

```python
def chatbot():
    print("\nWelcome to the Customer Support Chatbot! Type 'exit' to quit.\n")
    while True:
        user_query = input("You: ").lower()
        if user_query == "exit":
            print("Chatbot: Thank you for using our service!")
            break

        category, response = responses.get(user_query, (None, None))
        if response:
            print(f"Chatbot ({category}): {response}")
            log_query(user_query, category, response)
        else:
            default_response = "Sorry, I couldn't understand that. Please try again."
            print(f"Chatbot: {default_response}")
            log_query(user_query, "Unknown", default_response)
```

```python
log_file = "chatbot_logs.json"

def log_query(query, category, response):
    log_entry = {
        "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "query": query,
        "category": category,
        "response": response
    }

    try:
        with open(log_file, "r") as file:
            logs = json.load(file)
    except FileNotFoundError:
        logs = []

    logs.append(log_entry)
    with open(log_file, "w") as file:
        json.dump(logs, file, indent=4)
```

```python
def view_logs(filter_category=None, filter_date=None):
    try:
        with open(log_file, "r") as file:
            logs = json.load(file)
    except FileNotFoundError:
        print("No logs found.")
        return

    for log in logs:
        if (not filter_category or log["category"] == filter_category) and \
           (not filter_date or log["timestamp"].startswith(filter_date)):
            print(log)
```

```python
if __name__ == "__main__":
    chatbot()
```
